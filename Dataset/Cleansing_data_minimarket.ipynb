```python
# 1. Import library yang diperlukan
import pandas as pd
import re
import string

# 2. Baca dataset hasil crawling
df = pd.read_csv('Pelayanan_pelanggan_minimarket(Indomaret All).csv')

# 3. Simpan salinan kolom original untuk perbandingan
df['full_text_original'] = df['full_text']
```

```python
# 4. Fungsi cleaning
def clean_text(text):
    text = str(text)

    # Hapus URL
    text = re.sub(r'http\S+', '', text)

    # Hapus mention
    text = re.sub(r'@\w+', '', text)

    # Hapus hashtag
    text = re.sub(r'#\w+', '', text)

    # Hapus emoji dan karakter non-ASCII
    text = text.encode('ascii', 'ignore').decode('ascii')

    # Hapus karakter spesial dan angka
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # Ubah ke huruf kecil dan hapus spasi berlebih
    text = re.sub(r'\s+', ' ', text).strip().lower()

    return text
```

```python
# 5. Bersihkan isi tweet
df['full_text'] = df['full_text'].apply(clean_text)

# 6. Hapus baris dengan tweet kosong
df = df[df['full_text'].str.strip() != '']

# 7. Hapus duplikat berdasarkan isi tweet
df = df.drop_duplicates(subset='full_text')

# 8. Hapus tweet pendek (â‰¤ 3 kata)
df = df[df['full_text'].apply(lambda x: len(x.split()) > 3)]
```

```python
# 9. Simpan ke file baru
df.to_csv('Cleaned_Pelayanan_pelanggan_minimarket_Indomaret.csv', index=False)

# 10. Tampilkan ringkasan hasil
print(f"Jumlah tweet setelah dibersihkan: {len(df)}")
df[['full_text_original', 'full_text']].head()
```
